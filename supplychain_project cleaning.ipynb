**the next couple of blocks clean the data**

1) isnull() returns true/false if there is a missing for that feature
2) any() true if any is true and false if everything is false
3)axis=0 is rows and axis=1 is columns; so it's checking if any of the column has a missing in each row

```python
missing_values = data.isnull().any(axis=1)
print("rows with missing values are")
print(missing_values)
```

1) duplicated() returns true/false for each row
2) data[data.duplicated()] returns dataframe of what rows are missing

```python
duplicate_values = data[data.duplicated()]
print("duplicate rows:")
print(duplicate_values)
```

#if there is missing , use: data.dropna(axis=0, inplace=True) inplace means replace the original file
#if there is a duplicate, use: data.drop_duplicates(inplace=True)
