```python
data = pd.read_csv("supply_chain_data.csv")
```

```python
data.head()
```

```python
defect_rate_by_product_type = data.groupby("Product type")["Defect rates"].mean()
print("Defect rates by product type is:")
print(defect_rate_by_product_type)
```

```python
plt.figure(figsize=(10,6))
defect_rate_by_product_type.plot(kind="bar", color='skyblue')
plt.title("Defect rates by product type")
plt.xlabel("product type")
plt.ylabel("mean defect rate")
plt.xticks(rotation=45)
plt.show()
```
